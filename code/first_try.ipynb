```python
import requests, json

#getting the station id
station = input("Enter the station id ")  #8519483 

#setting a dictionary up 
#The keys are the names that we will put into the link for each data point 
#The ones are place holders for the data which we will recieve 
data = {"air_pressure": 1, "water_temperature": 1, "air_temperature": 1, "water_level": 1}

#This for loop will use the link with the given station id and given data value name
#The data that is found with this link will be set asthe value in the dictionary 
for item in data.keys():
    r = requests.get(f"https://api.tidesandcurrents.noaa.gov/api/prod//datagetter?&station={station}&date=latest&units=english&datum=MLLW&product={item}&time_zone=LST_LDT&format=json&application=NOS.COOPS.TAC.COOPSMAP&interval=")
    x = json.loads(r.text)
    data[item] = x['data'][0]['v']
    
#Here the link is set with the given station id
r = requests.get(f"https://tidesandcurrents.noaa.gov/cgi-bin/stationtideinfo.cgi?Stationid={station}&datum=MLLW&timezone=LST_LDT&units=english&clock=12hour&decimalPlaces=2")
tide_data = str(r.text)

#We split on each paragraph to get the data from each individual tide seprate 
tides = tide_data.split('\n')
    
#We split on the '|' to get each data point from each tide seprate 
if len(tides) == 4:
    tide1 = tides[0].split('|')
    tide2 = tides[1].split('|')
    tide3 = tides[2].split('|')
    tide4 = [0, 0] 
    next_tide = tides[3].split('|')

if len(tides) == 5:
    tide1 = tides[0].split('|')
    tide2 = tides[1].split('|')
    tide3 = tides[2].split('|')
    tide4 = tides[3].split('|')
    next_tide = tides[4].split('|')

print(data)

if tide1[0] == next_tide[0]:
    print("The next tide is: ")
    for item in tide1:
        print(item)
        
if tide2[0] == next_tide[0]:
    print("The next tide is: ")
    for item in tide2:
        print(item)
        
if tide3[0] == next_tide[0]:
    print("The next tide is: ")
    for item in tide3:
        print(item)

if tide4[0] == next_tide[0]:
    print("The next tide is: ")
    for item in tide4:
        print(item)
```
